```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
x = np.linspace(0,10,101)
y1=x**3
y2=x**2

plt.plot(x, y1, '--', label='x^3') 
plt.plot(x, y2, '-.', label='x^2') 
plt.title('График x в степени 2 и 3') 
plt.xlabel('ось X') 
plt.ylabel('ось Y') 
plt.legend()
plt.show()
```
